```python
# 3. Low Level Data Understanding :
#     a. Find count of unique values in location column.
#     b. Find which continent has maximum frequency using values counts.
#     c. Find maximum & mean value in 'total_cases'.
#     d. Find 25%,50% & 75% quartile value in 'total_deaths'.
#     e. Find which continent has maximum 'human_development_index'.
#     f. Find which continent has minimum 'gdp_per_capita'.
```

```python
# 3. a. Find count of unique values in location column.

# Solution :- 

# count unique values in 'location' column
unique_count = data['location'].nunique()

# print the count of unique values in 'location' column
print("The count of unique values in 'location' column is: ", unique_count)
```

```python
# 3. b. Find which continent has maximum frequency using values counts.

# Solution :- 

# get value counts for the 'continent' column
continent_counts = data['continent'].value_counts()

# get the index with the highest count
max_continent = continent_counts.idxmax()

# print the continent with the highest count
print("The continent with the highest frequency is: ", max_continent)
```

```python
# 3. c. Find maximum & mean value in 'total_cases'.

# Solution :-

# find maximum value in 'total_cases' column
max_cases = data['total_cases'].max()

# find mean value in 'total_cases' column
mean_cases = data['total_cases'].mean()

# print the maximum and mean values in 'total_cases' column
print("The maximum value in 'total_cases' column is: ", max_cases)
print("The mean value in 'total_cases' column is: ", mean_cases)
```

```python
# 3. d. Find 25%,50% & 75% quartile value in 'total_deaths'.

# Solution :-

# find quartile values in 'total_deaths' column
quartiles = data['total_deaths'].quantile([0.25, 0.5, 0.75])

# print the quartile values in 'total_deaths' column
print("The 25% quartile value in 'total_deaths' column is: ", quartiles[0.25])
print("The 50% quartile value in 'total_deaths' column is: ", quartiles[0.5])
print("The 75% quartile value in 'total_deaths' column is: ", quartiles[0.75])
```

```python
# 3. e. Find which continent has maximum 'human_development_index'.

# Solution :-

# group the DataFrame by continent and find maximum 'human_development_index'
max_hdi = data.groupby('continent')['human_development_index'].max()

# find the index (continent name) with the maximum value
max_continent = max_hdi.idxmax()

# print the continent with the maximum 'human_development_index'
print("The continent with the maximum 'human_development_index' is: ", max_continent)
```

```python
# 3. f. Find which continent has minimum 'gdp_per_capita'.

# Solution :- 

# group the DataFrame by continent and find minimum 'gdp_per_capita'
min_gdp = data.groupby('continent')['gdp_per_capita'].min()

# find the index (continent name) with the minimum value
min_continent = min_gdp.idxmin()

# print the continent with the minimum 'gdp_per_capita'
print("The continent with the minimum 'gdp_per_capita' is: ", min_continent)
```

```python
# 4. Filter the dataframe with only this columns
# ['continent','location','date','total_cases','total_deaths',
# 'gdp_per_capita','human_development_index'] and update the data frame.

# Solution :- 

# select the desired columns and update the DataFrame
Updated_data = data[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index'
                    ]]

# print the updated DataFrame
print(Updated_data)
```

```python
# 5. Data Cleaning
#     a. Remove all duplicates observations
#     b. Find missing values in all columns
#     c. Remove all observations where continent column value is missing
#     Tip : using subset parameter in dropna
#     d. Fill all missing values with 0
```

```python
# 5. a. Remove all duplicates observations

# Solution :- 

# drop all duplicates and update the DataFrame
df = Updated_data.drop_duplicates()

# print the updated DataFrame
print(df)
```

```python
# 5. b. Find missing values in all columns

# Solution :- 

# find missing values in all columns
missing_values = df.isnull().sum()

# print the missing values for all columns
print("Missing values in all columns:\n", missing_values)
```

```python
# 5. c. Remove all observations where continent column value is missing
#    Tip : using subset parameter in dropna

# Solution :-

# remove all observations where continent column value is missing
Updated_df = df.dropna(subset=['continent'])

# print the updated DataFrame
print(Updated_df)
```

```python
# 5. d. Fill all missing values with 0

# Solution :- 

# fill all missing values with 0
df = df.fillna(0)

# print the updated DataFrame
print(df)
```

```python
# 6. Date time format :
#     a. Convert date column in datetime format using pandas.to_datetime
#     b. Create new column month after extracting month data from date column.
```

```python
# 6. a. Convert date column in datetime format using pandas.to_datetime

# Solution :- 

# convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# print the updated DataFrame
print(df)
```

```python
# 6. b. Create new column month after extracting month data from date column.

# Solution :-

# convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# create new column month after extracting month data from date column
df['month'] = df['date'].dt.month

# print the updated DataFrame
print(df)
```

```python
# 7. Data Aggregation:
#     a. Find max value in all columns using groupby function on 'continent'column
#     Tip: use reset_index() after applying groupby
#     b. Store the result in a new dataframe named 'df_groupby'.
#     (Use df_groupby dataframe for all further analysis)
```

```python
# 7. a. Find max value in all columns using groupby function on 'continent'column
#     Tip: use reset_index() after applying groupby

# Solution :-

# group the DataFrame by continent and find the maximum value in each group
max_values = df.groupby('continent').max().reset_index()

# print the resulting DataFrame
print(max_values)
```

```python
# 7. b. Store the result in a new dataframe named 'df_groupby'.
#     (Use df_groupby dataframe for all further analysis)

# Solution :-

# group the DataFrame by continent and find the maximum value in each group
df_groupby = df.groupby('continent').max().reset_index()

# print the resulting DataFrame
print(df_groupby)
```

```python
# 8. Feature Engineering :
#     a. Create a new feature 'total_deaths_to_total_cases' by ratio of'total_deaths' column to 'total_cases'

# Solution :-

# create new column 'total_deaths_to_total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']

# print the updated DataFrame
print(df_groupby)
```
