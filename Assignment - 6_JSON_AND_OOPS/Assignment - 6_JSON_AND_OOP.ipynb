```python
# Assignment 1
# ðŸ‘‰ 1. Create a JSON file (employee.json) containing employee information of minimum 5 employees.
# Each employee information consists of Name, DOB, Height, City, State. Write a python program 
# that reads this information from the JSON file and saves the information into a list of objects of Employee class. 
# Finally print the list of the Employee objects.

import json

class Employee:
    def __init__(self):
        self.Employee_Details = {}
        self.Employee_Details_ID = len(self.Employee_Details) + 1
        print("Welcome to office")
    def Employee_Data(self):
        Employee_Details = {}
        for i in range(5):
            Employee_Name = input("Enter Employee Name :  ")
            Employee_DOB = input("Enter Employee DOB : ")
            Employee_Height = float(input("Enter Employee Height :  "))
            Employee_City = input("Enter Employee City :  ")
            Employee_State = input("Enter Employee State : ")
            Employee_Details = {"Employee Name" : Employee_Name , "Employee_DOB" : Employee_DOB , "Employee Height" : Employee_Height , "Employee City" : Employee_City , "Employee State" : Employee_State}
            self.Employee_Details_ID = len(self.Employee_Details) + 1
            self.Employee_Details[self.Employee_Details_ID] = Employee_Details
        with open("Employee.json", "w") as f:
            json.dump(self.Employee_Details , f)

    def Read_Emp_Data(self):
        with open("Employee.json", "r") as f:
            R = json.load(f)
            print(R)
            for k,v in R.items():
                print(k,v)
anu = Employee()
anu.Employee_Data()
anu.Read_Emp_Data()
```

```python
1# Assignment 2
# ðŸ‘‰ 1. Create a class named â€˜Dogâ€™. It should have a constructor which accepts its 
# name, age and coat color. You must perform the following operations:

# ðŸ”´ a. It should have a function â€˜description()â€™ which prints the name and age of the dog.
# ðŸ”´ b. It should have a function â€˜get_info()â€™ which prints the coat color of the dog.
# ðŸ”´ c. Create child classes â€˜JackRussellTerrierâ€™ and â€˜Bulldogâ€™ which is inherited 
# from the class â€˜Dogâ€™. It should have at least two methods of its own.
# ðŸ”´ d. Create objects and implement the above functionalities.

class Dog:
    
    def __init__(self,name,age,coat_color):
        self.age = age
        self.name = name
        self.coat_color = coat_color
        print("Welcome to Dog's Den")
    def description(self):
        return self.name , self.age
    def get_info(self):
        return self.coat_color
    
class JackRussellTerrier(Dog):
    
    def barks(self):
        barking = "bark "+" bark"
        return barking
    def uniques(self):
        fact = "It can jump high"
        return fact
    
class Bulldog(Dog):
    
    def LEGS(self):
        leg = "it has 4 legs"
        return leg
    def LOOKS(self):
        look = "It will look like a bull"
        return look
    

anu = Dog("snufy" , 2 , "black")
print(anu.description())
print(anu.get_info())

krish = JackRussellTerrier("hary" , 5 , "white")
print("child-1 class method called" , krish.uniques())
print("Parent Class Method called" , krish.description())
print("Parent Class Method called" , krish.get_info())
print(krish.barks())

A = Bulldog("ponnu" , 2 , "brown")
print("Parent Class Method called" , A.description())
print("Parent Class Method called" , A.get_info())
print("child-2 class method called" , A.LEGS())
print("child-2 class method called" , A.LOOKS())

print("This type of inheritance is called as Hierarchical inheritace which one parent class and multiple child classes")
```
